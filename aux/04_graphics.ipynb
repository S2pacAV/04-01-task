```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as st
```

##### 1. Bernoulli distr pmf

```python
p = 0.6

bernoulli_distr = st.bernoulli(p = p)
x = [0, 1]

fig, ax = plt.subplots(dpi=150)
ax.bar(x, bernoulli_distr.pmf(x), color='#599FFF', edgecolor='#4520ab', alpha=0.7)
ax.set_xticks([0, 1])
ax.set_xticklabels(['tail (0)', 'head (1)'])
ax.set_yticks([min(1 - p, p), max(1 - p, p)])
ax.set_xlabel(r'$k$')
ax.legend()
ax.set_ylabel(r'$PMF_X(k)$')
ax.set_title(r'$PMF$ of Coin flipping', size=20);
plt.savefig('../../slides/sources/aux/bernoulli_distr.jpg')
```

##### 2. Normal distr pdf

```python
normal_distr1 = st.norm(2, 0.5)
normal_distr2 = st.norm(4, 0.5)
normal_distr3 = st.norm(6, 2)

N = 10000

# sample1 = normal_distr.rvs(N)
```

```python
fig, ax = plt.subplots(figsize=(10, 5), dpi=170)

x = np.linspace(0, 12, 1000)

# sns.histplot(sample, ax=ax, color='#599FFF', stat='density')
ax.plot(x, normal_distr1.pdf(x), color='#599FFF', linewidth=1.5, label = r'PDF, $\mu=2, \sigma=0.5$')
ax.plot(x, normal_distr2.pdf(x), color='#4520ab', linewidth=1.5, label = r'PDF, $\mu=4, \sigma=0.5$')
ax.plot(x, normal_distr3.pdf(x), color='r', linewidth=1.5, label = r'PDF, $\mu=6, \sigma=2$')
ax.grid()
ax.set_xticks(np.arange(0, 12, 2))
ax.set_title(r'PDF of normal distribution - $\mathbb{N}(\mu, \sigma)$')

ax.legend();
plt.savefig('../../slides/sources/aux/normal_distr.jpg')
```

```python
dice_distr = st.randint(1, 7)

n = 2000

sample = np.array([])
means = np.array([])

for i in range(n):
    sample = np.append(sample, dice_distr.rvs(1))
    means = np.append(means, sample.mean())


fig, ax = plt.subplots(figsize=(15, 7), dpi=200)
ax.set_facecolor('w')
plt.plot(np.arange(1, n+1), means, color='#599FFF', linewidth=2.5)
plt.axhline(3.5, linestyle = ':', linewidth=3, color='#4520ab')
plt.xlabel('number of rolls (n)', size=20)
plt.ylabel(r'sample_mean ($\overline{X}$)', size=20)
plt.grid(color='grey')
plt.title('mean dice roll by number of rolls', size=20);
# plt.savefig('../../slides/sources/aux/law_of_large_numbers.jpg')
```

```python
p = 0.5
n = 10

bernoulli_distr = st.binom(n = n, p = p)
x = np.arange(n+1)

fig, ax = plt.subplots(dpi=150)
ax.bar(x, bernoulli_distr.pmf(x), color='#599FFF', edgecolor='r', alpha=0.7, label=r'$y \sim Bi(10, 0.5)$')
ax.set_xticks(x)
ax.set_yticks(np.round(bernoulli_distr.pmf(x), 3))
ax.set_xlabel(r'$k$')
ax.grid()
ax.legend()
ax.set_ylabel(r'$PMF_y(k)$')
ax.set_title(r'$PMF$ of the binomial distribution', size=20);
# plt.savefig('../../slides/sources/aux/binom_distr.jpg')
```

```python
_lambda1 = 7
_lambda2 = 3

pois_distr1 = st.poisson(_lambda1)
pois_distr2 = st.poisson(_lambda2)

x = np.arange(16)

fig, ax = plt.subplots(dpi=150)
ax.bar(x, pois_distr1.pmf(x), color='#599FFF', edgecolor='r', alpha=0.7, label=r'$y \sim Pois(\lambda=7)$')
ax.bar(x, pois_distr2.pmf(x), color='#4520ab', edgecolor='r', alpha=0.7, label=r'$y \sim Pois(\lambda=3)$')





ax.set_xticks(np.arange(16))
ax.set_xlabel(r'$k$')
ax.grid()
ax.legend()
ax.set_ylabel(r'$PMF_y(k)$')
ax.set_title(r'$PMF$ of poisson distribution', size=20);
plt.savefig('../../slides/sources/aux/pois_distr.jpg')
```

```python
_lambda1 = 7
_lambda2 = 3

expon_distr1 = st.expon(scale = _lambda1)
expon_distr2 = st.expon(scale = _lambda2)

x = np.arange(21)
x_continius = np.linspace(0, 21, 1000)

fig, ax = plt.subplots(dpi=150)
ax.plot(x_continius, expon_distr1.pdf(x_continius), color='#4520ab', alpha=0.7, label=r'$y \sim Expon(\lambda=7)$')
ax.bar(x, expon_distr1.pdf(x), color='#4520ab', edgecolor='r', alpha=0.7)


ax.plot(x_continius, expon_distr2.pdf(x_continius), color='#599FFF', alpha=0.7, label=r'$y \sim Expon(\lambda=3)$')
ax.bar(x, expon_distr2.pdf(x), color='#599FFF', edgecolor='r', alpha=0.7)


ax.set_xticks(np.arange(0, 21, 4))
# ax.set_yticks(np.round(expon_distr.pdf(x), 3))
ax.set_xlabel(r'$k$')
ax.grid()
ax.legend()
ax.set_ylabel(r'$PDF_y(k)$')
ax.set_title(r'$PDF$ of exponential distribution', size=15);
# plt.savefig('../../slides/sources/aux/expon_distr.jpg')
```

```python
distribution_of_purchased_coffee = st.binom(300, 0.5)

x = np.arange(118, 192)

plt.figure(figsize=(15, 8))

plt.plot(x, distribution_of_purchased_coffee.pmf(x), color='#4520ab')
plt.bar(x, distribution_of_purchased_coffee.pmf(x), color='#4520ab', label=r'$T(X) = \sum_{i=1}^{300}x_i \sim Bi(300, 0.5)$, если $H_0$ действительно верна')
plt.bar(167, distribution_of_purchased_coffee.pmf(167), color='r', label='Результат по нашей выборке')
plt.xlabel('Количество покупок кофе')
plt.ylabel('Вероятность получить такое количество покупателей')
plt.title(r'Распределение статистики(количество купивших людей) $T(X) = \sum_{i=1}^{300}x_i$', size=15)
plt.grid()
plt.xlim((118, 182))
plt.ylim((0, distribution_of_purchased_coffee.pmf(150)))
plt.xticks(np.arange(120, 185, 5))
plt.yticks(distribution_of_purchased_coffee.pmf([130, 135, 140, 145, 150]))
plt.legend()
plt.axvline(163.5, color='r', linestyle='-.', linewidth=2)



# plt.savefig('../../slides/sources/aux/binom_distr.png')
```

```python
distribution_of_purchased_coffee = st.binom(300, 0.5)

x = np.arange(118, 192)

plt.figure(figsize=(15, 8))

plt.plot(x, distribution_of_purchased_coffee.pmf(x), color='#4520ab')
plt.bar(x, distribution_of_purchased_coffee.pmf(x), color='#4520ab', label=r'$T(X) = \sum_{i=1}^{300}x_i \sim Bi(300, 0.5)$, если $H_0$ действительно верна')
plt.bar(167, distribution_of_purchased_coffee.pmf(167), color='r', label='Результат по нашей выборке')
plt.xlabel('Количество покупок кофе')
plt.ylabel('Вероятность получить такое количество покупателей')
plt.title(r'Распределение статистики(количество купивших людей) $T(X) = \sum_{i=1}^{300}x_i$', size=15)
plt.grid()
plt.xlim((118, 182))
plt.ylim((0, distribution_of_purchased_coffee.pmf(150)))
plt.xticks(np.arange(120, 185, 5))
plt.yticks(distribution_of_purchased_coffee.pmf([130, 135, 140, 145, 150]))
plt.legend()
plt.axvline(163.5, color='r', linestyle='-.', linewidth=2)


x = np.linspace(118, 182, 1000)
# Закрасить область справа от вертикальной линии
plt.fill_between(x, distribution_of_purchased_coffee.pmf(150), color='#599FFF', alpha=0.3, where=(x >= 163.5))

plt.fill_between(x, distribution_of_purchased_coffee.pmf(150), color='y', alpha=0.1, where=(x <= 163.5))

# Добавить текст для H0
plt.text(140, 0.03, 'H0', fontsize=32, color='black', ha='center')

# Добавить текст для H1
plt.text(174, 0.03, 'H1', fontsize=32, color='black', ha='center')
plt.tight_layout()

plt.savefig('../../slides/sources/aux/critical_binom_distr.png')
```
